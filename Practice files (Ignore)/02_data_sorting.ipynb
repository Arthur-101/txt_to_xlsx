```python
import re
import pandas as pd
```

```python
# Read the raw text file
with open('65027 XII.TXT.txt', 'r') as file:
    raw_data = file.read()
```

```python
# Regular expressions to extract relevant information
school_pattern = r"SCHOOL\s*:\s*-\s*(\d+)\s*(.+)"
date_pattern = r"DATE\s*:-\s*(\d{2}/\d{2}/\d{4})"
region_pattern = r"REGION:\s*([A-Z]+)"

main_data_pattern = r"\d+\s+[MF]\s+(.+?)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(.+)\s+(.+)\s+(.+)"
```

```python
# Extracting data using regular expressions
school_match = re.search(school_pattern, raw_data)
date_match = re.search(date_pattern, raw_data)
region_match = re.search(region_pattern, raw_data)
```

```python
# Extracting main data
main_data = re.findall(main_data_pattern, raw_data)
```

```python
# Prepare the extracted data into lists
school_code = school_match.group(1)
school_name = school_match.group(2)
exam_date = date_match.group(1)
region = region_match.group(1)
```

```python
rolls = []
sex = []
names = []
marks = []
grades = []
results = []

for roll, gender, name, *subject_marks, total_marks, result in main_data:
    rolls.append(roll.strip())
    sex.append(gender.strip())
    names.append(name.strip())
    marks.append(subject_marks)
    grades.append(total_marks.strip())
    results.append(result.strip())
```

```python
# Create a DataFrame
data = {
    'Roll': rolls,
    'Sex': sex,
    'Name of Candidate': names,
    'Subject Marks': marks,
    'Total Marks': grades,
    'Result': results,
}

df = pd.DataFrame(data)
df
```
